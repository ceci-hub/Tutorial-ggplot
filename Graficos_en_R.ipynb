## Se cambian variables categóricas a factor. 

Con el fin de realizar las estadisticas descriptivas moda, mediana, media


```r
df$Education <- as.factor(df$Education)
df$EnvironmentSatisfaction <- as.factor(df$EnvironmentSatisfaction)
df$JobSatisfaction <- as.factor(df$JobSatisfaction)
df$PerformanceRating <- as.factor(df$PerformanceRating)
df$RelationshipSatisfaction <- as.factor(df$RelationshipSatisfaction)
df$StockOptionLevel <- as.factor(df$StockOptionLevel)
df$OverTime <- as.factor(df$OverTime) 
df$MaritalStatus <- as.factor(df$MaritalStatus) 
df$Over18 <- as.factor(df$Over18)               
df$JobRole <- as.factor(df$JobRole)               
df$Gender <- as.factor(df$Gender)               
df$EducationField <- as.factor(df$EducationField)               
df$Department <- as.factor(df$Department)               
df$BusinessTravel <- as.factor(df$BusinessTravel)               
df$Attrition <- as.factor(df$Attrition)               
            

factores <- df %>%  select_if(is.factor)
names(factores)
noFactores <- df[,!names(df) %in% names(factores)]
```

# Estadisticas Descriptivas

```r
# Descriptivos
mean <- sapply(noFactores, mean) # Agrega  na.rm=TRUE para remover NA
sd <- sapply(noFactores, sd)
median <- sapply(noFactores, median)
max <- sapply(noFactores, max)
Nas <- colSums(is.na(noFactores))
length(df) # Tamaño
q1 <- t(sapply(noFactores, quantile)) #Cuantil Q1, hago la traspuesta con t
estadisticosNoFactores <- round(cbind(mean, median, sd, max, Nas, q1),digits=1) #cbind pasa a columna
estadisticosNoFactores
```

#Boxplot

```r
p1 <- ggplot(noFactores, aes("var", noFactores$MonthlyIncome)) +
  geom_boxplot(outlier.alpha = .25) +
  scale_y_log10(
    labels = scales::dollar, 
    breaks = quantile(noFactores$MonthlyIncome)
  )

p2 <- ggplot(noFactores, aes("var", noFactores$MonthlyIncome)) +
  geom_point() +
  geom_violin() +
  scale_y_log10(
    labels = scales::dollar, 
    breaks = quantile(noFactores$MonthlyIncome)
  )

gridExtra::grid.arrange(p1, p2, ncol = 2)
```

```r
# distribución de empleados por edad y género
y <- as.data.frame(table(df$edad, df$Gender))

ggplot(y, aes(x = Var1, y = Freq, fill = Var2)) + 
  # Defino genero
    geom_bar(data = subset(y, Var2 == "Female"), stat = "identity", mapping = aes(y = Freq), color="grey", size=0.2) + 
  geom_bar(data = subset(y, Var2 == "Male"), stat = "identity", position = "identity", mapping = aes(y = -Freq), color="grey", size=0.2) +
  # Defino escalas
  scale_y_continuous(breaks=seq(-51,51,4),labels=abs(seq(-51,51,4))) +
  # barplot horizontal
  coord_flip() +
  # Titulo
  theme_bw() +
  xlab("Edad (años)") + 
  ylab("Total de empleados") +
  ggtitle("Distribución de empleados por edad y género")

```
